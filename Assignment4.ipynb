```python
#Checks how many constraints the lists passes, and returns the number of constraints passed as well as the constraints not passed.
def constraints(list):
    result = 0
    result_list = []
    A = list[0]
    B = list[1]
    C = list[2]
    D = list[3]
    E = list[4]
    F = list[5]
    G = list[6]
    H = list[7]
    if A > G:
        result += 1
    else:
        result_list.append("A > G")
    if abs(G-C) == 1:
        result += 1
    else:
        result_list.append("|G - C| = 1")
    if D != C:
        result += 1
    else:
        result_list.append("D != C")
    if G != F:
        result += 1
    else:
        result_list.append("G != F")
    if abs(E-F) % 2 == 1:
        result += 1
    else:
        result_list.append("|E - F| is odd")
    if A <= H:
        result += 1
    else:
        result_list.append("A <= H")
    if abs(H-C) % 2 == 0:
        result += 1
    else:
        result_list.append("|H - C| is even")
    if E != C:
        result += 1
    else:
        result_list.append("E != C")
    if H != F:
        result += 1
    else:
        result_list.append("F != F")
    if abs(F-B) == 1:
        result += 1
    else:
        result_list.append("|F - B| = 1")
    if H != D:
        result += 1
    else:
        result_list.append("H != D")
    if E < (D-1):
        result += 1
    else:
        result_list.append("E < D - 1")
    if C != F:
        result += 1
    else:
        result_list.append("C != F")
    if G < H:
        result += 1
    else:
        result_list.append("G < H")
    if D >= G:
        result += 1
    else:
        result_list.append("D >= G")
    if E != (H-2):
        result += 1
    else:
        result_list.append("E != H - 2")
    if D != (F-1):
        result += 1
    else:
        result_list.append("D != F - 1")
        
    results = [result, result_list]
    return results
    #print(results[1])
```

```python
def print_pop(pop):
    current_population = 0
    for x in pop:
        current_population +=1
        print("Population", current_population, ":", x)
```

```python
def check_results(population):
    results = []
    current_population = 0
    for x in population:
        results.append(constraints(x))
    return results
```

```python
def print_results(results):
    current_population = 0
    for x in results:
        current_population +=1
        print("Population", current_population, ":", x[0])
```

```python
def probability(population):
    probability = []
    denominator = 0
    results = check_results(population)
    # get denominator
    for x in results:
        denominator += x[0]
        
    # calculate probability
    for x in results:
        probability.append(x[0] / denominator)
    return probability
```

```python
def print_prob(probability):
    current_population = 0
    for x in probability:
        current_population +=1
        print("Population", current_population, ":", x * 100)
```

```python
def selection(population, probability):
    result = random.choices(population,weights = probability, k = 8)
    return result
```

```python
def find_parents(old_population,new_population):
    temp = []
    for x in new_population:
        test_list1 = x.copy()
        for y in old_population:
            test_list2 = y.copy()
            if test_list1 == test_list2:
                temp.append(x) 
    temp.sort()
    output = list(temp for temp,_ in itertools.groupby(temp))
    return output
```

```python
def print_selected_population(population):
    counter = 0
    for x in population:
        counter += 1
        print(counter, ":",x)
```

```python
def pick_random_numbers():
    result = []
    for x in range(4):
        n = random.randrange(1, 7)
        result.append(n)
    return result
```

```python
def crossover(population,random_crossover):
    new_population = []
    for x in random_crossover:
        newlist = population[0][0:x] + population[1][x:8]
        newlist2 = population[1][0:x] + population[0][x:8]
        population.remove(population[0])
        population.remove(population[0])
        new_population.append(newlist)
        new_population.append(newlist2)
    return new_population
```

```python
def print_crossover(results):
    current_population = 0
    for x in results:
        current_population +=1
        print("Population", current_population, ":", x)
```

```python
def mutation(population):
    mutation_list = []
    mutated_pop = []
    mutated_summary = []
    result = []
    # create a list consisting of [to mutate(1), or to not mutate (0),
    #                              which index to mutate,
    #                              which number to mutate to]
    a = random.choices([0,1], weights=(0.7,0.3), k=8)
    b = random.choices([0,1,2,3,4,5,6,7], weights = [1,1,1,1,1,1,1,1],k=8)
    c = random.choices([1,2,3,4],weights = [1,1,1,1], k = 8)
    mutation_list.append(a)
    mutation_list.append(b)
    mutation_list.append(c)
    
    for x in range(8):
        if mutation_list[0][x] == 1:
            pop = population[x].copy()
            pop[mutation_list[1][x]] = mutation_list[2][x]
            mutated_pop.append(pop)
            mutated_summary.append([x,mutation_list[1][x],population[x][mutation_list[1][x]],mutation_list[2][x]])
        else:
            mutated_pop.append(population[x])
    result.append(mutated_pop)
    result.append(mutated_summary)
    return result
```

```python
def print_mutation(mutation_summary):
    result = mutation_summary[0]
    summary = mutation_summary[1]
    current_population = 0
    for y in summary:
        print("population",y[0],": changed at index",y[1],',',y[2],"->",y[3])
    for x in range(8):
        current_population +=1
        print("Population", current_population, ":", result[x])
```

```python
import random
import itertools
def genetic_alg(initial_population, n, show):
    #set up constants
    initial_pop = initial_population
    Max_fitness = 17
    all_population = [initial_population]

    if show:
        for x in range(n):
            print("---------------------------------------------------------------------")
            print("Generation", x)
            print("---------------------------------------------------------------------")
            initial_pop = all_population[x]
            print('Populations:')
            print_pop(initial_pop)
            print("---------------------------------------------------------------------")

            # gets the number of constraints passed and fitness
            results = check_results(initial_pop)
            print("Fitness: (Max = 17)")
            print_results(results)
            print("---------------------------------------------------------------------")
        
            # calculates the probability of being chosen based on fitness score
            population_probability = probability(initial_population)
            print("Population Probability:")
            print_prob(population_probability)
            print("---------------------------------------------------------------------")
            
            # selects the new population
            selected_population = selection(initial_pop, population_probability)
            print("Selected Populations:")
            parents = find_parents(initial_pop,selected_population)
            print("Parents:",parents)
            print_selected_population(selected_population)
            print("---------------------------------------------------------------------")
            
            # preforms crossover
            random_crossover = pick_random_numbers()
            print("Crossover:")
            print('Crossover occurring at index',random_crossover[0], "for population 1 and 2")
            print('Crossover occurring at index',random_crossover[1], "for population 3 and 4")
            print('Crossover occurring at index',random_crossover[2], "for population 5 and 6")
            print('Crossover occurring at index',random_crossover[3], "for population 7 and 8")
            new_population = crossover(selected_population, random_crossover)
            print("Crossover population:")
            print_crossover(new_population)
            print("---------------------------------------------------------------------")
            
            # mutates the populations with a probability of 30%
            mutated_population = mutation(new_population)
            print('Mutations:')
            print("Mutated_population:")
            print_mutation(mutated_population)
            all_population.append(mutated_population[0])
    else:
        for x in range(n):
            initial_pop = all_population[x]

            # gets the number of constraints passed and fitness
            results = check_results(initial_pop)
        
            # calculates the probability of being chosen based on fitness score
            population_probability = probability(initial_population)
            
            # selects the new population
            selected_population = selection(initial_pop, population_probability)
            parents = find_parents(initial_pop,selected_population)
            
            # preforms crossover
            random_crossover = pick_random_numbers()
            new_population = crossover(selected_population, random_crossover)
            
            # mutates the populations with a probability of 30%
            mutated_population = mutation(new_population)
            all_population.append(mutated_population[0])
            
    return all_population
```
